```python
#import pandas and read in the excel file along with each of its sheets in its respective dataframe

import pandas as pd
import tldextract

data = pd.ExcelFile('Forbes Assessment Sept. 2022.xlsx')
clicks = pd.read_excel(data, 'Clicks on Forbes Links')
thirdparty = pd.read_excel(data, '3rdParty Alt Retailers')
retailer_epc = pd.read_excel(data, 'Retailer EPCs')

pd.set_option('display.max_rows', 1000)
```

```python
high_epc = df_cmb9[['url', 'Product Name', 'Retailer', 
                          'EPC','Alt Retailer Names', 'Alt EPC', 
                          'Clicks']]
high_epc
```

```python
high_epc2 = high_epc.groupby(['Product Name'])['EPC', 'Alt EPC', 'Clicks'].max().reset_index()

#high_epc2['Alt EPC'] = high_epc.groupby(['Product Name'])['Alt EPC'].max()

#high_epc2['Alt EPC'].fillna(0)

high_epc2['Alt EPC'] = high_epc2['Alt EPC'].fillna(value=0)

high_epc2
```

```python
import numpy as np

high_epc2['Current Earnings'] = high_epc2['EPC'] * high_epc2['Clicks']
high_epc2['Alt Earnings'] = high_epc2['Alt EPC'] * high_epc2['Clicks']
high_epc2['Earnings % Diff'] = (high_epc2['Alt Earnings'] - high_epc2['Current Earnings']) / high_epc2['Current Earnings'] * 100

high_epc3 = high_epc2.sort_values('Earnings % Diff', ascending = False)


high_epc3['Alt Earnings2'] = high_epc3.loc[high_epc3['Earnings % Diff'] <= 0, 'Alt Earnings'] = high_epc2['Current Earnings']
#high_epc3

high_epc3 = high_epc3.drop(['Alt Earnings2', 'Earnings % Diff'], axis=1)
high_epc3
```

```python
high_epc3['Total Current Earnings'] = high_epc3['Current Earnings'].sum()

high_epc3['Total Alt Earnings'] = high_epc3['Alt Earnings'].sum()

high_epc3['Estimated Earnings'] =  high_epc3['Total Alt Earnings'] - high_epc3['Total Current Earnings']

high_epc3['Estimated Lift %'] = (high_epc3['Total Alt Earnings'] - high_epc3['Total Current Earnings'])/ high_epc3['Total Current Earnings'] * 100

high_epc3
```

```python
retailers = dict(zip(retailer_epc.EPC, retailer_epc.Domain))

high_epc3['Higher Alt Retailer'] = high_epc3['Alt EPC'].map(retailers)

high_epc3['Original Retailer'] = high_epc3['EPC'].map(retailers)

high_epc3
```
